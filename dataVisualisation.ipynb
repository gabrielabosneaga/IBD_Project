```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def renameColumns(df):
    return [col.split('-', 1)[1] if '-' in col else col for col in df.columns]
```

```python
def convertTime(df):
    df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S')
    return df
```

```python
def isWeekend(dateObj):
    return dateObj.weekday() in [5, 6]
```

```python
df_08_45_18 = pd.read_csv("sensorData/08_45_18.csv")
df_08_45_18.columns = renameColumns(df_08_45_18)
df_08_45_18 = convertTime(df_08_45_18)
df_08_45_18.head()
```

```python
#Separate df with data only from weekdays
weekendMask = df_08_45_18['Time'].apply(isWeekend)
df_08_45_18_weekdays = df_08_45_18[~weekendMask]
df_08_45_18_weekdays = df_08_45_18_weekdays.reset_index(drop=True)
df_08_45_18_weekdays
```

```python
def makeDfSensorSingle(df, sensorName):
    dfSensor = pd.DataFrame()
    sensorColumns = ['Time'] + [col for col in df.columns if sensorName in col] 
    dfSensor = df[sensorColumns]
    dfSensor = dfSensor.dropna(axis=1, how='all')
    dfSensor = dfSensor.dropna().reset_index(drop=True)
    return dfSensor

def makeDfSensorType(df, sensorName):
    dfSensor = pd.DataFrame()
    sensorColumns = ['Time'] + [col for col in df.columns if sensorName in col] 
    dfSensor = df[sensorColumns]
    dfSensor = dfSensor.dropna(how='all')
    dfSensor = dfSensor.dropna(axis=1, how='all')
    return dfSensor
```
